# Club Penguin Dance Contest Automation:
In this project, we focus on automating a perfect run on club penguin dance contest. To do this we will generally use the following algorithm:
1. Use canny edge detected on a color filtered image to identify the game canvas
2. Identify the rows that contain the empty arrows
3. Indentify if and what type of arrow is in the zone
4. Use this to press buttons in real time from a screen recording

### First we will import necessary libraries:

```python
import cv2
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import glob
```

### We next import an image of the blank board:

```python
test_img = cv2.imread('data/testimg_empty.png')

plt.imshow(cv2.cvtColor(test_img, cv2.COLOR_BGR2RGB))
```

### We then filter it to only pixels with a blue channel value greater than our threshold:

```python
def filter_blue(blue_value, red_green_sum, threshold_blue = 240, threshold_red_blue = 200, output_value = 255):
    if (blue_value >= threshold_blue):
        if (red_green_sum <= threshold_red_blue):
            return [output_value, output_value, output_value]
    return [0,0,0]

blue = [[filter_blue(pixel[0], pixel[1]+pixel[2]) for pixel in row] for row in test_img]
plt.imshow(blue)
print(blue)
```

### We then use canny edge detection on this new image to identify the game canvas:

```python
edges = cv2.Canny(blue,100,200)
plt.imshow(edges)
```
